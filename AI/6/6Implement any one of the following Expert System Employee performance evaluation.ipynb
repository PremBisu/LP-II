Employee performance evaluation

```python
import matplotlib.pyplot as plt

def evaluate_performance(punctuality, task_completion, teamwork, innovation):
    scores = {}
    total = 0

    # Rule 1: Punctuality
    if punctuality.lower() == "good":
        scores["Punctuality"] = 2
    elif punctuality.lower() == "average":
        scores["Punctuality"] = 1
    else:
        scores["Punctuality"] = 0

    # Rule 2: Task Completion
    if task_completion.lower() == "excellent":
        scores["Task Completion"] = 3
    elif task_completion.lower() == "good":
        scores["Task Completion"] = 2
    elif task_completion.lower() == "average":
        scores["Task Completion"] = 1
    else:
        scores["Task Completion"] = 0

    # Rule 3: Teamwork
    if teamwork.lower() == "excellent":
        scores["Teamwork"] = 3
    elif teamwork.lower() == "good":
        scores["Teamwork"] = 2
    elif teamwork.lower() == "average":
        scores["Teamwork"] = 1
    else:
        scores["Teamwork"] = 0

    # Rule 4: Innovation
    if innovation.lower() == "high":
        scores["Innovation"] = 3
    elif innovation.lower() == "moderate":
        scores["Innovation"] = 2
    elif innovation.lower() == "low":
        scores["Innovation"] = 1
    else:
        scores["Innovation"] = 0

    total = sum(scores.values())

    # Final evaluation
    if total >= 10:
        evaluation = "Outstanding"
    elif total >= 7:
        evaluation = "Good"
    elif total >= 4:
        evaluation = "Satisfactory"
    else:
        evaluation = "Needs Improvement"

    return scores, total, evaluation

def visualize(scores, total, evaluation, employee_name):
    categories = list(scores.keys())
    values = list(scores.values())

    plt.figure(figsize=(8, 4))
    bars = plt.bar(categories, values, color='orange', edgecolor='black')

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.1, int(yval), ha='center', va='bottom')

    plt.ylim(0, 4)
    plt.title(f"{employee_name} - {evaluation} (Score: {total}/11)")
    plt.ylabel("Score (out of 3)")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# --- Main Execution ---
print("=== Employee Performance Evaluation Expert System ===")
n = int(input("Enter the number of employees: "))

for i in range(n):
    print(f"\n--- Employee {i+1} ---")
    employee_name = input("Enter employee name: ")
    punctuality = input("Enter punctuality (Good/Average/Poor): ")
    task_completion = input("Enter task completion (Excellent/Good/Average/Poor): ")
    teamwork = input("Enter teamwork (Excellent/Good/Average/Poor): ")
    innovation = input("Enter innovation (High/Moderate/Low/None): ")

    scores, total, evaluation = evaluate_performance(punctuality, task_completion, teamwork, innovation)

    print(f"\n{employee_name}'s Final Evaluation: {evaluation}")
    print(f"Score Breakdown: {scores}")

    visualize(scores, total, evaluation, employee_name)
```

________________________________________
 Theory: Employee Performance Evaluation Expert System (Implemented in Python)
 Objective:
The aim of this expert system is to automate the evaluation of employees' performance using a rule-based approach. It collects inputs for various performance parameters, scores them using predefined rules, and visualizes the results using a bar chart.
________________________________________
 Evaluation Criteria:
The system evaluates each employee based on four key parameters:
1.	Punctuality (Good/Average/Poor)
2.	Task Completion (Excellent/Good/Average/Poor)
3.	Teamwork (Excellent/Good/Average/Poor)
4.	Innovation (High/Moderate/Low/None)
Each parameter is scored as follows:
	Punctuality: Max 2 points
	Task Completion, Teamwork, Innovation: Max 3 points each
	Total max score = 11
________________________________________
 Working:
1.	Input: User enters qualitative responses for the four parameters.
2.	Processing:
o	Inputs are converted into numerical scores using if-else rules.
o	The total score is computed.
o	Final evaluation is based on the total:
	Outstanding (1011)
	Good (79)
	Satisfactory (46)
	Needs Improvement (<4)
3.	Output: The system prints the employee's evaluation and shows a bar chart of parameter-wise scores using matplotlib.

 Features:
	Multi-employee evaluation support.
	User-friendly interaction via command-line input.
	Visual insights using score-based bar charts.

 Benefits:
